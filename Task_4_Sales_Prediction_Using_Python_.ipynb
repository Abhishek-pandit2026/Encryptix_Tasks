```python
# Remove outliers from the 'Newspaper' column
df = df[df['Newspaper'] < 100]
```

```python
# Performing univariate analysis and creating visualizations
plt.rcParams['figure.figsize'] = (18, 4)
plt.subplot(1, 3, 1)
sns.distplot(df['TV'], color='cadetblue')
plt.subplot(1, 3, 2)
sns.distplot(df['Radio'], color='seagreen')
plt.subplot(1, 3, 3)
sns.distplot(df['Newspaper'], color='green')
plt.subplot(1, 3, 1)
sns.distplot(df['Sales'], color='crimson')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['TV']], df[['Sales']], test_size = 0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
res= model.predict(X_test)
print(res)
```

```python
# Assuming you have X_test, y_test, and model already defined

# Get predictions
y_pred = model.predict(X_test)

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print metrics
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"R-squared Score: {r2:.2f}")

# Plot actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted Sales")
plt.show()

# Residual analysis
residuals = y_test - y_pred
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals, color='blue', alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel("Predicted Sales")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()

# Distribution of residuals
plt.figure(figsize=(10, 6))
plt.hist(residuals, bins=20, edgecolor='black')
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Distribution of Residuals")
plt.show()

# Interpretation
print("\nInterpretation:")
print(f"1. The model explains {r2:.2%} of the variance in the target variable.")
print(f"2. On average, the model's predictions are off by {mae:.2f} units (Mean Absolute Error).")
print(f"3. The Root Mean Squared Error of {rmse:.2f} gives the standard deviation of the residuals.")
print("4. Check the 'Actual vs Predicted Sales' plot to visualize how well the predictions match actual values.")
print("5. The Feature Importance plot shows which features have the strongest impact on predictions.")
print("6. In the Residual Plot, look for any patterns. Ideally, residuals should be randomly scattered around zero.")
print("7. The Distribution of Residuals should ideally be normal and centered around zero.")
```
